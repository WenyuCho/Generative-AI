風格 : **普普風**

普普藝術（Pop Art） 起源於1950～60年代，是一種結合大眾文化與藝術的視覺風格。

代表人物包括 Andy Warhol、Roy Lichtenstein。

普普風就是把日常生活中常見的東西，畫成像漫畫一樣又亮又誇張的圖畫。


把平凡的東西畫得像漫畫、像廣告、像流行海報，就是普普風。
