**作業目標** :

 Python 結合 Google Colab ，實現一個**正切函數 (tangent)** 的畫圖，


主要透過使用 **numpy** 生成數據點、**matplotlib** 繪圖，

再對參數進行微調，達到最終結果。

**步驟1**
      
引入必要套件



```python
import numpy as np #用於數學運算
import matplotlib.pyplot as plt #用於繪製圖形
```

**步驟2**
      
進行圖形繪製，參數調整測試。

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500) # 設定x範圍 ，分成500個點，點數太少會不夠平滑
y = np.tan(x) # 計算對應的 y 值，也就是 tan(x)

y = np.clip(y, -10, 10) # 使用 np.clip 函數來限制 y 值範圍，y值範圍過大會導致圖形失真

# 繪製圖形
plt.plot(x, y, label='y = tan(x)', color='orange') #線的顏色變橘色
plt.title('Tangent Function') #圖表標題

#設定x、y軸標籤
plt.xlabel('x')
plt.ylabel('y')

# 設定 x 和 y 軸的基準線（y=0 和 x=0）
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

plt.grid(True) # 顯示網格線，幫助更清楚地查看圖形
plt.legend()
#plt.show() # 在colab無需使用此語法也會show
```
